```python
import pandas as pd

# Load your downloaded file
df = pd.read_csv('Kenyan_Motor_Insurance_2023_2024.csv')
```

```python

# 2. Calculate the Loss Ratio 
df['Loss_Ratio'] = df['Total_Claim_Amount_KES'] / df['Annual_Premium_KES']
```

```python

# 3. Add the Expense Ratio (Industry standard benchmark of 30%)
df['Combined_Ratio'] = df['Loss_Ratio'] + 0.30
```

```python

# 3. Create 'Underwriting Status'
df['Underwriting_Status'] = df['Combined_Ratio'].apply(
    lambda x: 'Profitable' if x < 1.0 else 'Loss Making'
)
```

```python

# 4. Create a 'Risk Profile'
def categorize_risk(row):
    if row['Total_Claim_Amount_KES'] == 0:
        return 'Low Risk'
    elif row['Total_Claim_Amount_KES'] <= row['Annual_Premium_KES']:
        return 'Moderate Risk'
    else:
        return 'High Risk'

df['Risk_Profile'] = df.apply(categorize_risk, axis=1)
```

```python

# 5. Fill missing Accident Causes
df['Accident_Cause'] = df['Accident_Cause'].fillna('No Claim')
```

```python

# Save as a generic professional file
df.to_csv('Kenya_Insurance_Analysis_Cleaned.csv', index=False)
print("Step 1 Complete: File saved as 'Kenya_Insurance_Analysis_Cleaned.csv'")
```
